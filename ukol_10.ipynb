```python
import pandas as pd
import matplotlib.pyplot as plt
platy = pd.read_csv('platy_2021_02.csv')
print(platy)
```

```python

bin_edges = range(0, int(platy['plat'].max()) + 5000, 5000)

plt.hist(platy['plat'], bins=bin_edges, edgecolor='black')

plt.title('Rozložení platů v softwarové firmě (únor 2021)')
plt.xlabel('Plat')
plt.ylabel('Počet zaměstnanců')

plt.show()
```
